## Описание проекта

Требуется построить модель МО в интересах золотодобывающей компании.\
Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды.\
Предоставлены данные как с параметрами очистки, так и добычи.\
Имеется 3 дата-сета:
- Обучающая выборка. Содержит набор данных для обучения.
- Тестовая выборка. Не содержит целевые прихнаки.
- Исходные данные. Содержит полный набор данных, но с пропусками.

Учитывая формат и содержание предоставленных данных:
- Будем использовать кросс-валидацию.
- Понадобится предобработка данных.

Технологический процесс
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- Xanthate — ксантогенат (промотер, или активатор флотации);
- Sulphate — сульфат (на данном производстве сульфид натрия);
- Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота
- Параметры этапов
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

Наименование признаков имеет вид:
[этап].[тип_параметра].[название_параметра]

- Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Возможные значения для блока [тип_параметра]:
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

## Пользовательские функции

```python
# Функция для первичного анализа дата-сета
def analyze_df(df):
    # выводим размерность датафрейма
    display("Размерность датафрейма:")
    display(df.shape)
    display("")

    # выводим информацию о датафрейме
    display("Информация о датафрейме:")
    display(df.info())
    display("")

    # выводим статистическую информацию о датафрейме
    display("Статистическая информация о датафрейме:")
    display(df.describe())
    display("")

    # выводим первые 10 строк датафрейма
    display("Первые 10 строк датафрейма:")
    display(df.head(10))

    # выодим кол-во пропусков
    display("Кол-во пропусков")
    display(df.isna().sum()[df.isna().sum() > 0])
    
    # выодим кол-во дубликатов
    display("Кол-во дубликатов")
    display(df.duplicated().sum()[df.duplicated().sum() > 0])
```

```python
# Функция расчета эффективности
def output_recovery(C, F, T):
    recovery = (C * (F - T)) / (F * (C - T)) * 100
    return recovery

# C — доля золота в концентрате после флотации/очистки;
# F — доля золота в сырье/концентрате до флотации/очистки;
# T — доля золота в отвальных хвостах после флотации/очистки
```

### Проверим, что эффективность обогащения рассчитана правильно.

```python
# выведем необходимые расчеты в отдельный датафрейм
test_recovery = train[['rougher.output.concentrate_au', 
                            'rougher.input.feed_au', 
                            'rougher.output.tail_au', 
                            'rougher.output.recovery']]

# рассчитаем значение
test_recovery['check_recovery'] = test_recovery.apply(
    lambda row: output_recovery(row['rougher.output.concentrate_au'], 
                                row['rougher.input.feed_au'], 
                                row['rougher.output.tail_au']), 
    axis=1)

# посмотрим таблицу
test_recovery.head()
```

```python
# разделим на предсказания и целевой результат
test_recovery_target = test_recovery['rougher.output.recovery']
test_recovery_predictions = test_recovery['check_recovery']

# рассчитаем MAE
round(mean_absolute_error(
    test_recovery_target, 
    test_recovery_predictions
), 2)
```

Значение среднего абсолютного отклонения = 0, значит расчеты выполнены верно

## Модель

### Функция вычисления итоговой sMAPE

```python
def smape(y_true,y_pred):
    numerator = np.abs(y_true - y_pred)
    denominator = (np.abs(y_true) + np.abs(y_pred))/2
    smape = ((numerator/denominator)*100).sum() / len(y_pred)
    return 0.25*smape[0] + 0.75*smape[1]
```

```python
features = train.drop(columns_not_needed, axis=1)
target = train[['rougher.output.recovery', 'final.output.recovery']]

scaler = StandardScaler()
scaler.fit(features) 
features = scaler.transform(features)
```

### Решающее дерево

```python
best_depth_t = 0
final_score_b = 1000

for depth in tqdm(range(2, 12)):
    dt = DecisionTreeRegressor(random_state=12345, max_depth=depth)
    scores = -1 * cross_val_score(dt, features, target, scoring=smape_scorer, cv=5)
    final_score = scores.mean()
    if final_score < final_score_b:
        final_score_b = final_score
        best_depth_t = depth
        
# сохраним лучшие параметры в словаре
tree_best = {'depth':best_depth_t}
print('Средняя оценка качества модели:', final_score_b)
```

```python
# Определить словарь параметров
param_dist = {'max_depth': range(2, 12)}

# Определить модель
dt_radomCV = DecisionTreeRegressor(random_state=12345)

smape_scorer = make_scorer(smape, greater_is_better=False)

# Определить объект RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=dt_radomCV,
                                   param_distributions=param_dist,
                                   n_iter=10,
                                   scoring=smape_scorer,
                                   cv=5,
                                   random_state=12345)

# Обучить модель
random_search.fit(features, target)

# Получить лучшие параметры
best_params = random_search.best_params_
scores = -1 * cross_val_score(random_search, features, target, scoring=smape_scorer, cv=5)
final_score_dt = scores.mean()

# Распечатать результаты
print(f"Лучшие параметры: {best_params}")
print(f"Лучшая оценка: {final_score_dt}")
```
